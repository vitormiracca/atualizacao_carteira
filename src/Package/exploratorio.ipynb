```python
def obter_cotacao_crypto(symbol):
    exchange = ccxt.binance()  # Pode ser substituído por outra exchange

    try:
        ticker = exchange.fetch_ticker(symbol)
        return ticker['close']
    except ccxt.NetworkError as e:
        print(f"Erro de rede: {e}")
    except ccxt.ExchangeError as e:
        print(f"Erro da exchange: {e}")
    except Exception as e:
        print(f"Erro desconhecido: {e}")

# Exemplo de uso
symbol = 'BTC/BRL'  # Substitua pela criptomoeda desejada e a moeda de cotação desejada
cotacao = obter_cotacao_crypto(symbol)
print(f"A cotação de {symbol} é: {cotacao}")
```
